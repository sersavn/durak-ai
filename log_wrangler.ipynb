### Log wrangler

```python
import json
import pandas as pd

transformed_result = None
path = 'game.log'
log_data = open(path, 'r')

json_data = []
for i in log_data:
    if i[0] == '{':
        i = i.replace("'", "\"")
        j = json.loads(i)
        json_data.append(j)
        
transformed_data = []
for i in json_data:
    if list(i.keys())[0] == 'Game':
        curr_game = i
    elif list(i.keys())[0] == 'Round':
        curr_round = i
    else:
        d = {**curr_game, **curr_round, **i}
        transformed_data.append(d)
```

```python
df = pd.DataFrame(transformed_data)
df = df.set_index(['Game', 'Round'])
df
```
