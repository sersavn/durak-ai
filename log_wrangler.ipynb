### log wrangler 1.0

```python
import pandas as pd
transformed_result = None
path = 'game.log'
log_data = open(path, 'r')

transformed_result = []
for i in log_data:
    if i[0] == 'G':
        curr_game = i.rstrip()
    elif i[0] == 'R':
        curr_rnd = i.rstrip()
    else:
        transformed_result.append([curr_game.split('-')[1], curr_rnd.split('-')[1], i.rstrip()])
        # Game_Num, Round_Num, Actions_Log
```

```python
print(transformed_result)
```

```python
# Initial dataframe
df = pd.DataFrame(transformed_result, columns = ['Game_Num', 'Round_Num', 'Actions_Log'])
df
```

```python
# Transformed dataframe
df = df.set_index(['Game_Num', 'Round_Num'])
```
